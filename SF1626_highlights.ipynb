# 12.2 Limits and Continuity


These are very similar to the notions for functions of one variable.

The limit of $f(x,y)$ as $(x,y)$ approaches $(a,b)$ is denoted as $$\lim_{(x,y)\rightarrow (a,b)} f(x,y)$$

It is essential to remember and udnerstand Theorem 2 in the book.

#### Definition of the limit of $f(x,y)$ at $(a,b)$
We consider a function $f$ defined in a neighborhood of the poiint $(a,b)$.
$$\lim_{(x,y)\rightarrow(a,b)} f(x,y) = L $$ 
provided that
   * for every $\epsilon>0$, there exists $\delta(\epsilon)$ such that 
   * $ |f(x,y)-L|<\epsilon$ **whenver** $\sqrt{(x-a)^2+(y-b)^2}<\delta(\epsilon)$ 
        (i.e. **whenever $(x,y)$ is within the disk of radius $\delta(\epsilon)$, centered at $(a,b)$**)

#### Definition of continuity

Definition 3: $f(x,y)$ is continuous at $(a,b)$ if
   * $\lim_{(x,y)\rightarrow(a,b)} f(x,y)$ exists (call it $L$), and
   * $L=f(a,b)$

# Partial Derivatives

* The partial derivative $f_1(a,b)$ is **the rate of change of $f(x,y)$ at $x=a$ while $y=b$**; i.e. it is the rate of change of $f$ along the line $(x,y=b)$. 
* We can use our knowledge of calculus for functions of one variables: $f_1(a,b) = \frac{d}{dx} f(x, b)|_{x=a} $

* Similarly, the partial derivative $f_2(a,b)$ is the rate of change of $f(x,b)$ at $x=a$.

Of course, it is always good to remember the definitions of these partial derivatives as a limiting process.

Along the curve: $t\mapsto (a+t,b)$ the rate of change at $(a,b)$ is defined formally as

$$f_1(a,b):=\lim_{h\rightarrow 0} \frac{ f(a+h,b)-f(a,b)}{h}$$

Along the curve: $t\mapsto (a,b+t)$ the rate of change at $(a,b)$ is defined formally as

$$f_2(a,b):=\lim_{h\rightarrow 0} \frac{ f(a,b+h)-f(a,b)}{h}$$

## Directional Derivatives

Let $\mathbf{u}=u~\mathbf{i}+ v~\mathbf{j}$ be a unit vector; i.e. $u^2+v^2=1$.

**Definition 7**: The directional derivative of $f$ at $(a,b)$ is **the rate of change of $f$ along the line/ray in the direction $\mathbf{u}$ which passes through $(a,b)$**

* The ray can be traced by $(a+tu, b+tv)$ for $-\infty<t<\infty$.

$$D_\mathbf{u} f(a,b) := \frac{d}{dt} f(a+tu, b+tv) = f_1(a,b)u+f_2(a,b)v = \nabla f(a,b)\cdot \mathbf{u}$$

In the formula above, you need to know how the directional derivatives of $f$ is related to the gradient of $f$.

* In the formula above, **it is important that $\mathbf{u}$ is a unit vector!** (Why?)

Directional derivative: 

$$D_\mathbf{u} f = \nabla f \cdot \mathbf{u} = |\nabla f||\mathbf{u}|\cos\theta$$

* $\theta$ is the angle between $\nabla f$ and $\mathbf{u}$.
* The maximal value of $\cos\theta$ is $1$ when $\theta=0,\pm 2\pi,\cdots$ (when $\mathbf{u}$ and $\nabla f$ point to the **same direction**)
* The minimal value of $\cos\theta$ is $-1$ (when $\mathbf{u}$ and $\nabla f$ point to the **opposite direction**)
* $\cos\theta = 0$ when $\mathbf{u}\perp \nabla f$ (The rate of change of $f$ in the directions **perpendicular** to $\nabla f$ is $0$)

### Tangent planes and normal lines (of the graph of $f$)

* You need to know the defintions
* You need to know how to find the normal vectors of the graph of $f$
* You need to be able to derive the equations of the tangent planes and normal lines


#### normal vectors to the graph of $f$

* Cut the graph of $f$ by the plane $y=b$. The curve at the intersection can be written parametrically as $x\mapsto(x, b, f(x,b))$. The **tangent vector** at $(a,b)$: 

$$T_1:= \mathbf{i}+ f_1(a,b)\mathbf{k}$$

* Cut the graph of $f$ by the plane $x=a$. The curve at the intersection can be written parametrically as $y\mapsto(a, y, f(a,y))$. The **tangent vector** at $(a,b)$:

$$T_2:= \mathbf{j}+ f_2(a,b)\mathbf{k}$$

* a normal vector $n$ to the graph of $f$ can be found by taking the cross product of $T_3$ and $T_1$:
$$\mathbf{n}:=T_2\times T_1=\text{det}\left(\begin{array}{ccc}
\mathbf{i} & \mathbf{j} & \mathbf{k}\\
0 & 1 & f_{2}\\
1 & 0 & f_{1}
\end{array}\right) = f_1\mathbf{i}+f_2\mathbf{j}-\mathbf{k}$$

Reminder: $\mathbf{n}=\mathbf{n}(x,y)$ is a vector if $\mathbb{R}^3$.

#### Getting the normal vectors from the gradient

Let's start with finding the normal vector of the graph of a fucntion of one variable: $f(x)$.

* Define a new function of two variable by $g(x,y)=y-f(x)$.
* The 0-level set of $g$ is teh graph of $f$.
* Since $\nabla g (x,y) = (-f^\prime(x), 1)$ is perpendicular to the level sets of $g$, $\nabla g( a, f(a))$ is perpendicular to the graph of $f$ at $(a, f(a))$.
* A unit normal vector to the graph of $f$ as a function of $x$ is then $$ \mathbf{n}(x) = \frac{1}{\sqrt{f^\prime(x)^2+1}} (f^\prime(x)~\mathbf{i}-\mathbf{j})$$

* This generates to functions of two variables very easily.

#### Parametric definition of the normal line

Once we have the normal vector, the normal line can be defined easily: $$t\mapsto a~\mathbf{i}+b~\mathbf{j}+f(a,b)~\mathbf{k}+\mathbf{n}(a,b)~t$$

#### Equation for the tangent plane

Given the normal vector, $\mathbf{n}=(n_1,n_2,n_3)$, of a plane passing through $(a,b,c)$, we can quickly write down the equation for that plane:

$$ n_1(x-a)+n_2(y-b)+n_3(z-c)=0$$

(Every vector on the plane, i.e. $(x-a,y-b,z-c)$, should be orthogonal to $\mathbf{n}$)

In our particular case, 

$$ c=f(a,b) $$

$$\mathbf{n}=(-f_1(a,b), -f_2(a,b), 1) \implies z=f_1(a,b)(x-a)+f_2(a,b)(y-b)+f(a,b)$$

# Higher-Order Derivatives

$f(x,y)$ is a function mapping $(x,y)$ to a real number.

Its partial derivatives $f_1(x,y)$, $f_2(x,y)$ are also functions mapping $(x,y)$ to real numbers.

**So one can define the partial derivatives of $f_1$ and $f_2$ in exactly the same way.**

We have

$$f_{11}(x,y)\equiv\frac{\partial}{\partial x} f_1(x,y) = \lim_{h\rightarrow 0} \frac{f_1(x+h,y)-f_1(,x,y)}{h}$$

How would you define $f_{12}(x,y)$ and $f_{21}(x,y)$?

#### The mixed derivatives:

Partial derivatives of order 2: 

$$f_{21}(x,y)\equiv\frac{\partial}{\partial y} f_1(x,y) = \lim_{h\rightarrow 0} \frac{f_1(x,y+h)-f_1(,x,y)}{h}$$

For funcitons of two variables, very often we write:

$$f_{xx}\equiv f_{11},~~~f_{xy}\equiv f_{12},~~~etc.$$

#### Equivalence of partial derivatives: **Theorem 1 (p.698)  (VERY IMPORTANT!)**

It says that $f_{xy} (a,b) = f_{yx}(a,b)$ **if** $f_x, f_y, f_{xy}$ and $f_{yx}$ are all **continuous at $(a,b)$**


You now know how to define higher order partial derivatives of $f$:

$$f_{xxx}, f_{yxx},~etc.$$


**Theorem 1(p.698):**

$$ f_{xyx}(a,b)=f_{xxy}(a,b)=f_{yxx}(a,b) $$

if all the partial derivatives of $f$ of order $\le 3$ are continuous at $(a,b)$

## Linear Approximations

The tangent plane, $L(x,y)$ of $f$ at $(a,b)$ as a function is called the **linearization** of $f$ at $a$.

Again, to get the equation for the tangent plane:
* Get the normal vector of $f$ at $(a,b)$: $\mathbf{n} := f_1(a,b)\mathbf{i}+f_2(a,b)\mathbf{j}-\mathbf{k}$
* Define the inner product: 

$$ (x-a, y-b, z-f(a,b))^\top \cdot \mathbf{n}=0$$

$$\iff$$

$$ L(x,y):=z=f(a,b)+f_1(a,b)(x-a)+f_2(a,b)(y-b)$$

$L(x,y)$ is an affine function. $L(x,y)-f(a,b)$ is a linear function of $x$ and $y$.

What's the point of this?

* $L(a+h,b+y) \approx f(a,b)$: In general, $L$ is a very good approximation of $f$ near $(a,b)$
* Later, if $f$ can be approximated "well-enough" by $L$, we say that $f$ is differentiable.

```julia
#Example 1. $f(x,y)=\sqrt{2x^2+e^2y}$

xgrid = [ x for x in range(-0.5, stop=0.5, length=601), y in range(-0.5, stop=0.5, length=601) ]
ygrid = [ y for x in range(-0.5, stop=0.5, length=601), y in range(-0.5, stop=0.5, length=601) ]

f(x,y)=sqrt(2*x^2+exp(2y))

#get the partial derivatives of f
f1(x,y)=2x/f(x,y)
f2(x,y)=exp(2y)/f(x,y)

#the point around which linearization of f is to be performed
a,b=0.15,0.25

#write out the expression for the tangent plane L(x,y)
L(x,y)=f(a,b)+f1(a,b)*(x-a)+f2(a,b)*(y-b)

surf(xgrid, ygrid, f.(xgrid, ygrid),alpha=0.3)
surf(xgrid, ygrid, L.(xgrid, ygrid), alpha=0.75)

scatter3D( a,b, f(a,b))
```

```julia
surf(xgrid, ygrid, f.(xgrid, ygrid),alpha=0.3)
surf(xgrid, ygrid, L.(xgrid, ygrid), alpha=0.75)

scatter3D( a,b, f(a,b))

s=range(-0.75, stop=0.5, length=101)

τ=rand(2); τ./norm(τ)

#create a straight line passing through (a,b, f(a,b))
c1=a.+τ[1]*s
c2=b.+τ[2]*s
fval=f.(c1,c2)
Lval=L.(c1,c2)

plot3D(c1,c2, fval, "b")

plot3D(c1,c2, Lval, "r")
```

```julia
plot(s, Lval-fval)
title("The difference between L and f goes to 0 faster than any straight line")
```

### Differentiability

Definition 5 (p 714). 

Let $L(x,y)$ be the **linearization** of $f$ at $(a,b)$. 

We say that $f(x,y)$ is **differentiable** at the point $(a,b)$ if

$$ \lim_{(h,k)\rightarrow (0,0)}\frac{L(x+h,y+k)-f(a,b)}{\sqrt{h^2+k^2}}=0$$

This means that the difference between $L(x+h,y+k)$ and $f(a,b)$ approaches zero "**faster**" than $\sqrt{h^2+k^2}$ approaches zero, as $(h,k)\rightarrow(0,0)$.

E.g. $h^2+k^2$ goes to $0$ faster than $\sqrt{h^2+k^2}$.

#### A mean value theorem

Theorem 3 (p. 714)

If $f_1(x,y)$ and $f_2(x,y)$ are continuous in a neighborhood of the point $(a,b)$, and

If $|h|$ and $|k|$ are sufficiently small, then

there exist $\theta_1$ and $\theta_2$ between $0$ and $1$ such that

$$f(a+h,b+k) - f(a,b) = h f_1(a+\theta_1 h,b+k)+kf_2(a, b+\theta_2 k)$$ 

#### Theorem 4 (p.714) If $f_1(x,y)$ and $f_2(x,y)$ are continuous in a neighborhood of $(a,b)$, then $f$ is differentiable at $(a,b)$.

### Differentials

The differential of $f(x,y)$ is defined as $df:\mathbb{R}^2\times\mathbb{R}^2\mapsto \mathbb{R}$:

$$df\equiv df(x,y, dx, dy):= f_1(x,y)dx+f_2(x,y)dy$$

Compare $df$ to the linearization of $f$ around $(x,y)$: 

$$L(x+dx, y+dy):= f(x,y)+ f_1(x,y)dx+f_2(x,y)dy$$

It approximates the change of $f$. 

### General formulas for $\mathbf{f}:\mathbb{R}^n \mapsto \mathbb{R}^m$


$$\begin{align}y_{1}&=f(x_{1},x_{2},\cdots,x_{n})\\
y_{2}&=f_{2}(x_{1},x_{2},\cdots,x_{n})\\
&\vdots\\
y_{m}&=f_{m}(x_{1},x_{2},\cdots,x_{n})
\end{align}$$

The **Jacobian matrix** of $\mathbf{f}$:

$$D\mathbf{f}(\mathbf{x}):=\left(\frac{\partial f_{i}}{\partial x_{j}}(\mathbf{x})\right)\equiv\left(\begin{array}{cccc}
\frac{\partial f_{1}}{\partial x_{1}} & \frac{\partial f_{1}}{\partial x_{2}} & \cdots & \frac{\partial f_{1}}{\partial x_{n}}\\
\frac{\partial f_{2}}{\partial x_{1}} & \frac{\partial f_{2}}{\partial x_{2}} & \cdots & \frac{\partial f_{2}}{\partial x_{n}}\\
\\
\frac{\partial f_{m}}{\partial x_{1}} & \frac{\partial f_{m}}{\partial x_{2}} & \cdots & \frac{\partial f_{m}}{\partial x_{n}}
\end{array}\right)$$

### The differential of $\mathbf{f}$

$$ d\mathbf{y} := D\mathbf{f}(\mathbf{x})d\mathbf{x} $$

### Chain rule

$\mathbf{g}:\mathbb{R}^m \mapsto \mathbb{R}^k$

$$\begin{align}
z_{1}&=g_{1}(y_{1},y_{2},\cdots,y_{m})\\z_{2}&=g_{2}(y_{1},y_{2},\cdots,y_{m})\\&\vdots\\z_{k}&=g_{k}(y_{1},y_{2},\cdots,y_{m})
\end{align}$$

$$D(\mathbf{g\circ}\mathbf{f})(\mathbf{x})=D\mathbf{g}(\mathbf{f}(\mathbf{x)})D\mathbf{f}(\mathbf{x})$$
